```python
import numpy as np
import matplotlib.pyplot as plt
import os

main_folder = os.getcwd()

fig1 = plt.figure(1)
ax1 = fig1.add_subplot(111)

folders = ['run_detailed','run_detailed']
colors = ['tab:blue','tab:orange']
timesteps = [0,5000,10000,25000,30000]

for i, folder in enumerate(folders):
    for timestep in timesteps:
        s = np.load(main_folder + '/' + folder + '/profile_' + str(timestep) + 'yrs_s.npy')
        eta = np.load(main_folder + '/' + folder + '/profile_' + str(timestep) + 'yrs_eta.npy')
        if i == 0:
            ax1.plot(s,eta,color=colors[i],alpha=0.5)
        else:
            ax1.plot(s,eta/2.,color=colors[i],alpha=0.5)
ax1.set_xlabel('streamwise distance [m]')
ax1.set_ylabel('elevation [m]')
plt.show()
```
