```python
import numpy as np
import matplotlib.pyplot as plt
import os
import matplotlib.pylab as pl
import matplotlib

matplotlib.rcParams['mathtext.default'] = 'regular'
matplotlib.rcParams['font.sans-serif'] = "Trebuchet MS"
matplotlib.rcParams['font.family'] = "sans-serif"
matplotlib.rcParams['font.size'] = 12

main_folder = os.getcwd()

fig, axs = plt.subplots(2, 2)
fig.set_size_inches(8, 4)

folders = ['run_one','run_two','run_three','run_four']
colors = ['tab:blue','tab:orange','tab:green','tab:purple']
labels = ['reduce uplift', 'introduce soluble rock', 'change in climate', 'change in rock erodibility']
timesteps = [0,1000,5000,10000,20000,40000,100000]

colors_ii = pl.cm.copper_r(np.linspace(0,1,len(timesteps)))

sub_x = [0,0,1,1]
sub_y = [0,1,0,1]

for i, folder in enumerate(folders):
    for j, timestep in enumerate(timesteps):
        s = np.load(main_folder + '/clayandmarco/' + folder + '/profile_' + str(timestep) + 'yrs_s.npy')
        eta = np.load(main_folder + '/clayandmarco/' + folder + '/profile_' + str(timestep) + 'yrs_eta.npy')
        axs[sub_x[i],sub_y[i]].plot(s,eta,color=colors_ii[j],alpha=0.5)
        axs[sub_x[i],sub_y[i]].text(0,75,labels[i],ha='left',va='bottom')
axs[0,0].set_ylabel('elevation [m]')
axs[1,0].set_ylabel('elevation [m]')
axs[1,0].set_xlabel('streamwise distance [m]')
axs[1,1].set_xlabel('streamwise distance [m]')
plt.tight_layout()
plt.savefig('comparison.png',dpi=300)
```
